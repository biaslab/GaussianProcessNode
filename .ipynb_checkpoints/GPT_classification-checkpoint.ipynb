```julia
Random.seed!(145)
xmin, xmax = -5, 5  # Bounds of the data
N          = 100 # Number of samples
xtrain     = sort(rand(Uniform(xmin, xmax), N))
xtest      = collect(range(xmin, xmax; length=400))
kernelfunc =  2*with_lengthscale(SEKernel(),1.2)
meanfunc   = (x) -> 0.0;
grammatrix = kernelmatrix(kernelfunc,xtrain,xtrain) + 1e-6*diageye(N)
ytrain     = rand(MvNormal(zeros(N),grammatrix))
p          = normcdf.(ytrain)
ytrainbin  = map(rand,Bernoulli.(p));
ytrainbin = float(ytrainbin);
```

```julia
M = 15; #number of inducing points 
Xu = range(-5.,5.;length=M);
mean_gp = (x) -> 0
softplus = (x) -> log(1 + exp(x))
kernel_gp() = softplus([1]) * with_lengthscale(SEKernel(),softplus([2]))
gpcache = GPCache()
;
```

```julia
@model function gp_classification(y,x,)
    v ~ MvNormalMeanCovariance(zeros(M), 50diageye(M))
    w ~ GammaShapeRate(0.01,0.01)
    for i in eachindex(y)
        f[i] ~ UniSGP(x[i],v,w,) 
        y[i] ~ Probit(f[i])
    end
end

gp_constraints = @constraints begin
    q(f,v,w) = q(f)q(v)q(w)
end

@meta function meta_gp_classification(Xu,kernel,cache)
    UniSGP() -> UniSGPMeta(nothing,Xu,kernel,cache)
    Probit() -> ProbitMeta(21)
end

gp_initialization = @initialization begin
    q(v) = MvNormalMeanCovariance(zeros(M), 50diageye(M))
    q(w) = GammaShapeRate(1e-2,1e-2)
end
;
```

## Do inference

```julia
function my_free_energy()
    Kuu = kernelmatrix(kernel_gp(), Xu) + 1e-8 * I
    KuuL = fastcholesky(Kuu).L
    gpcache.cache_LowerTriangular[(:KuuL,M)] = KuuL
    infer_result = infer(
        model = gp_classification( = ),
        iterations = 10,
        data = (y = ytrainbin, x = xtrain,),
        initialization = gp_initialization,
        constraints = gp_constraints,
        returnvars = (v = KeepLast(),w = KeepLast(),f = KeepLast(),),
        meta = meta_gp_classification(Xu,kernel_gp,gpcache),
        free_energy = true,
    )
    return (infer_result.free_energy[end],infer_result.posteriors[:v], infer_result.posteriors[:w],infer_result.posteriors[:f])
end

function PerformInference(; epochs)
    FE_value = []
    q_v = []
    q_w = []
    _optimal = 
    @inbounds for epoch=1:epochs
            #step 1: Perform inference for v, w 
            fe, qv,qw, qf = my_free_energy(_optimal)
            #step 2: optimize the hyperparameters 
            _v, _v = mean_cov(qv)
            Rv = _v + _v * _v' |> (x) -> fastcholesky!(x).L
            w = mean(qw)
            f = mean.(qf)

            negllh = () -> neg_log_backwardmess_fast(;y_data=f, x_data=xtrain,v = _v,Lv=Rv, w=w,kernel=kernel_gp,Xu=Xu)
            grad! = (G, ) -> grad_llh_new!(G,  ;y_data=f, x_data=xtrain,v = _v,Lv=Rv, w=w,kernel=kernel_gp,Xu=Xu)
            res = optimize(negllh, grad!, _optimal, LBFGS())
            append!(FE_value,fe)
            push!(q_v, qv)
            push!(q_w,qw)
            _optimal = res.minimizer
    end
    return q_v, q_w, FE_value, _optimal
end
```

```julia
_init = [1.,1.]
qv,qw, FE_value, _optimal = PerformInference(_init;epochs=250);
```

```julia
function predict_new(x_test,qv,qw, q, meta)
    prediction_f = @call_rule UniSGP(:out, Marginalisation) (q_in=PointMass(x_test),q_v = qv, q_w = qw,q_ = q, meta=meta)
    prediction_y = @call_rule Probit(:out, Marginalisation) (m_in=prediction_f,meta=ProbitMeta(32))
    return prediction_f, prediction_y
end
```

```julia
predict_mean =[]
predict_var =[]
predict_func=[]
predict_var_func=[]
for i=1:length(xtest)
    prediction_f, prediction_y = predict_new(xtest[i],qv,qw,PointMass(_optimal), UniSGPMeta(nothing,Xu,kernel_gp,gpcache))
    append!(predict_func,mean(prediction_f))
    append!(predict_var_func,var(prediction_f)) 
    append!(predict_mean,mean(prediction_y))
    append!(predict_var,var(prediction_y)) 
end
```

```julia
function predictgp(x_test,xdata,ydata,kernel)
    Ktrain = kernelmatrix(kernel,xdata,xdata) + 1e-6*I
    k_cross = kernelmatrix(kernel,x_test,xdata)

    mean_f = k_cross * inv(Ktrain) * ydata 
end
ytest = predictgp(xtest,xtrain,ytrain,kernelfunc)
p_test          = normcdf.(ytest)
ytestbin  = map(rand,Bernoulli.(p_test));
ytestbin = float(ytestbin);
```
