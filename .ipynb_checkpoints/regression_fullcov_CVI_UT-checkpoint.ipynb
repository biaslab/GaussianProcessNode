```julia
using Revise 
using StableRNGs, GraphPPL,ReactiveMP, RxInfer, Random, Distributions, LinearAlgebra, Plots
using Flux, Zygote, ForwardDiff
using BenchmarkTools
import KernelFunctions: Matern52Kernel, with_lengthscale, Kernel 
import Functors: @functor
```

```julia
struct MyMatern52Kernel <: Kernel 
    l :: Real 
    σ :: Real 
end 
@functor MyMatern52Kernel
(k::MyMatern52Kernel)(x,y) = k.σ * (1 + sqrt(5)*abs(x-y)/k.l + 5*(x-y)^2/(3 * k.l^2)) * exp(-sqrt(5) * abs(x-y)/k.l)
```

```julia
Random.seed!(2)
xmin, xmax = -3, 3  # Bounds of the data
N = 100# Number of samples
xtrain = sort(rand(Uniform(xmin, xmax), N))
xtest = collect(range(xmin - 0.1, xmax + 0.1; length=200));
# xtest = sort(vcat(xtest,xtrain))
precision = 100
ytrain = sinc.(xtrain) + randn(N) * sqrt(1/precision);
f_true = sinc.(xtest);

k_func(θ) =  with_lengthscale(Matern52Kernel(),θ)
mean_func   = (x) -> 0.0;
θ_init = 4.0 #our initial lengthscale
log_θ_init = log(θ_init)

cov_strategy = CovarianceMatrixStrategy(FullCovarianceStrategy())  # 10 is the number of inducing points

# Random.seed!(2)
# xmin, xmax = -3, 3  # Bounds of the data
# N = 100# Number of samples
# xtrain = sort(rand(Uniform(xmin, xmax), N))
# xtest = collect(range(xmin - 0.1, xmax + 0.1; length=200));
# # xtest = sort(vcat(xtest,xtrain))
# precision = 100
# ytrain = sinc.(xtrain) + randn(N) * sqrt(1/precision);
# f_true = sinc.(xtest);

# k_func(θ) =  MyMatern52Kernel(θ[1],θ[2])
# mean_func   = (x) -> 0.0;
# θ_init = [2.0, 4.0] #our initial lengthscale
# log_θ_init = log.(θ_init)

# cov_strategy = CovarianceMatrixStrategy(SoR(50))  # 10 is the number of inducing points
```

```julia
# test new thing
@model function gpprocess(n,kernelfunc,meanfunc,train,test,strategy)
    y = datavar(Float64,n)
    x = randomprocess(test,train,strategy)
    ### change here 
    θ ~ NormalMeanVariance(log_θ_init, 2.)
    # x ~ GaussianProcess(meanfunc,kernelfunc, θ) where {meta = CVI(StableRNG(42), 1000,100,ADAM(0.1),ForwardDiffGrad(),10,Val(true),true), 
    #                     pipeline = RequireMarginal(params)}
    x ~ GaussianProcess(meanfunc,kernelfunc,θ) where {meta = Unscented(2,0, 1,nothing),pipeline = RequireMarginal(params)}
    #######
    γ ~ GammaShapeRate(.01,0.01)
    for i=1:n 
        y[i] ~ NormalMeanPrecision(x, γ) where {meta=ProcessMeta(i)}
    end
end

# @model function gpprocess(n,kernelfunc,meanfunc,train,test,strategy)
#     y = datavar(Float64,n)
#     x = randomprocess(test,train,strategy)
#     ### change here 
#     θ ~ MvNormalMeanCovariance(log_θ_init, [2. 0.5;0.5 2.])
#     x ~ GaussianProcess(meanfunc,kernelfunc, θ) where {meta = Unscented(2,0.,1,nothing),pipeline = RequireMarginal(params)}
#     #######
#     γ ~ GammaShapeRate(.01,0.01)
#     for i=1:n 
#         y[i] ~ NormalMeanPrecision(x, γ) where {meta=ProcessMeta(i)}
#     end
# end
```

```julia
@constraints function gpconstraints()   
    q(x,γ,θ) = q(x)q(γ)q(θ)
end
```

```julia
nits = 50
iresult = inference(
    model = gpprocess(length(ytrain), k_func, mean_func, xtrain,xtest, cov_strategy),
    initmarginals = (γ = GammaShapeRate(0.01,0.01),θ = NormalMeanVariance(log_θ_init, 2.)),
    constraints = gpconstraints(),
    iterations = nits,  
    data  = (y = ytrain,),
    free_energy = true
)

# iresult = inference(
#     model = gpprocess(length(ytrain), k_func, mean_func, xtrain,xtest, cov_strategy),
#     initmarginals = (γ = GammaShapeRate(0.01,0.01),θ = MvNormalMeanCovariance(log_θ_init, [2. 0.5;0.5 2.])),
#     constraints = gpconstraints(),
#     iterations = nits,  
#     data  = (y = ytrain,),
#     free_energy = true,
#     showprogress = true
# )
```

## some remarks
The experiment works under the following conditions:
- kernel: Matern52
- autodiff tool: Zygote, but use Zygote for 1st derivative, ForwardDiff for the second derivative 
- small dataset (e.g. 50-70 training data, 100 test set)
- use logdet(), don't use log(det())
