## Split to train and test set

```julia
N_train = 4000;
xtrain, ytrain = x_data[1:N_train], label[1:N_train]
xtest, ytest = x_data[N_train+1:end], label[N_train + 1: end];
```

## Split to batches

```julia
function split2batch(data::Tuple,batch_size)
    x, y = data
    x_batch = [x[i:min(i + batch_size - 1, end)] for i in 1:batch_size:length(x)]
    y_batch = [y[i:min(i + batch_size - 1, end)] for i in 1:batch_size:length(y)]
    return x_batch, y_batch
end

#create batch data
data_training = (xtrain, ytrain);
batch_size = 100;
xtrain_minibatches, ytrain_minibatches = split2batch(data_training, batch_size);
nbatches = length(ytrain_minibatches);
```

```julia
M = 300; #number of inducing points 
Random.seed!(1223)
#build a MvNormal for input
mean_x = mean(xtrain, dims = 1)
approx_cov = cov(xtrain)
dist = MvNormal(mean_x[1], approx_cov);

# Draw samples for inducing points
Xu = [rand(dist) for i=1:M];

kernel_gp(θ) = θ[1] * with_lengthscale(SEKernel(),[θ[2], θ[3]])
ndims_params = 3
θ_init = ones(ndims_params)
;
```

## Model

```julia
@model function gp_classification(y,x,θ)
    v ~ MvNormalMeanCovariance(zeros(M), 50diageye(M))
    w ~ GammaShapeRate(0.01,0.01)
    for i in eachindex(y)
        f[i] ~ UniSGP(x[i],v,w,θ) 
        y[i] ~ Probit(f[i])
    end
end

gp_constraints = @constraints begin
    q(f,v,w) = q(f)q(v)q(w)
end

@meta function meta_gp_classification(Xu,kernel)
    UniSGP() -> UniSGPMeta(nothing,Xu,kernel)
    Probit() -> ProbitMeta(30)
end

gp_initialization = @initialization begin
    q(v) = MvNormalMeanCovariance(zeros(M), 50diageye(M))
    q(w) = GammaShapeRate(1e-2,1e-2)
end
;
```

## Inference

```julia
Kuu_inverse = (θ) -> cholinv(kernelmatrix(kernel_gp(θ),Xu) .+ 1e-8)

Ψ_0 = (θ;xbatch) -> diag(kernelmatrix(kernel_gp(θ), xbatch))

Ψ_1 = (θ; xbatch) -> begin
    K = kernelmatrix(kernel_gp(θ), xbatch, Xu)
    return K[1:batch_size, :]
end

Ψ_2 = (θ; xbatch) -> begin
    K_Xu = kernelmatrix(kernel_gp(θ), Xu, xbatch)  # Compute kernel between Xu and all points in xtrain_minibatches[1]
    K_xi = kernelmatrix(kernel_gp(θ), xbatch)  # Compute kernel between all points

    # Use a comprehension or matrix operation to compute the product
    return [K_Xu[:, i] * K_Xu[:, i]' for i in 1:batch_size]  # Efficient calculation
end

function log_backwardmess(Rv_minus_Kuu_inv)
    return  w * μ_y[i] * (Ψ_1_values * μ_v)[1] - 
                    0.5 * w * (Ψ_0_values - tr(Ψ_2_values * Rv_minus_Kuu_inv))
end
# function grad_func()
```

```julia
function my_free_energy(θ; xbatch, ybatch)
    infer_result = infer(
        model = gp_classification(θ = θ),
        iterations = 20,
        data = (y = ybatch, x = xbatch,),
        initialization = gp_initialization,
        constraints = gp_constraints,
        returnvars = (v = KeepLast(),f = KeepLast(),w = KeepLast(),),
        meta = meta_gp_classification(Xu,kernel_gp),
        free_energy = false,
    )
    return (infer_result.posteriors[:v], infer_result.posteriors[:f], infer_result.posteriors[:w])
end

#for 1 batch
function PerformInference(θ;epochs = 1)
    # FE_value = []
    q_v = []
    grad = similar(θ)
    optimizer = Flux.AdaMax()
    θ_optimal = θ
    for epoch=1:epochs
        for b=1:nbatches
            #step 1: Perform inference for v, w 
            qv, qf, qw = my_free_energy(θ_optimal; xbatch = xtrain_minibatches[b], ybatch = ytrain_minibatches[b])
            #step 2: optimize the hyperparameters 
            μ_v, Σ_v = mean_cov(qv)
            Rv = Σ_v + μ_v * μ_v'
            μ_y = mean.(qf)
            w = mean(qw)

            neg_mllh = (θ) -> begin
            # Precompute values that don't change inside the loop
            Ψ_1_values = Ψ_1(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Ψ_0_values = Ψ_0(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Ψ_2_values = Ψ_2(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Kuu_inv = Kuu_inverse(θ)  # Compute once
            Rv_minus_Kuu_inv = Rv - Kuu_inv  # Precompute matrix subtraction
            # Sum over the batch size
            return -sum([
                        w * μ_y[i] * (Ψ_1_values[i] * μ_v)[1] - 
                        0.5 * w * (Ψ_0_values[i] - tr(Ψ_2_values[i] * Rv_minus_Kuu_inv))
                        for i = 1:batch_size
                    ])
            end

            ForwardDiff.gradient!(grad,neg_mllh, θ_optimal)
            Flux.Optimise.update!(optimizer,θ_optimal,grad)
            # append!(FE_value,fe)
            push!(q_v, qv)
        end
    end
    return q_v, θ_optimal
end
```

```julia
@time qv,θ_opt = PerformInference(θ_init; epochs = 1); # forwarddiff
```

```julia
@time qv,θ_opt = PerformInference(θ_init; epochs = 1); # zygote
```
