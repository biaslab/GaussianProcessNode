```julia
xtrain_data = matopen("data/kin40k/kin40k_xtrain.mat");
xtest_data = matopen("data/kin40k/kin40k_xtest.mat");
ytrain_data = matopen("data/kin40k/kin40k_ytrain.mat");
ytest_data = matopen("data/kin40k/kin40k_ytest.mat");

xtrain_matrix = read(xtrain_data, "xtrain");
ytrain = read(ytrain_data, "ytrain") |> (x) -> vcat(x...);
xtest_matrix = read(xtest_data, "xtest");
ytest = read(ytest_data, "ytest") |> (x) -> vcat(x...);

Ntrain = length(ytrain);
xtrain = [xtrain_matrix[i,:] for i=1:Ntrain];
data_training = (xtrain, ytrain);

Ntest = length(ytest);
xtest = [xtest_matrix[i,:] for i=1:Ntest];


;
```

```julia
function split2batch(data::Tuple,batch_size)
    x, y = data
    x_batch = [x[i:min(i + batch_size - 1, end)] for i in 1:batch_size:length(x)]
    y_batch = [y[i:min(i + batch_size - 1, end)] for i in 1:batch_size:length(y)]
    return x_batch, y_batch
end
```

```julia
#create batch data
batch_size = 10;
xtrain_minibatches, ytrain_minibatches = split2batch(data_training, batch_size);
nbatches = length(ytrain_minibatches);
```

```julia
Random.seed!(1)
M = 500;
pos = randperm(Ntrain)[1:M]
Xu = xtrain[pos]; #inducing input 

kernel_gp(θ) = θ[1] * with_lengthscale(SEKernel(),θ[2:end])
dim_θ = size(xtrain_matrix,2) + 1
θ_init = ones(dim_θ);
```

```julia
@model function gp_kin40k(y,x,θ)
    v ~ MvNormalMeanCovariance(zeros(M), 50diageye(M))
    for i in eachindex(y)
        y[i] ~ UniSGP(x[i],v,1e-3,θ) 
    end
end
```

```julia
@meta function meta_gp_regression(Xu,kernel)
    UniSGP() -> UniSGPMeta(nothing,Xu,kernel)
end
```

## Normal approach

```julia
function my_free_energy(θ; xbatch, ybatch)
    infer_result = infer(
        model = gp_kin40k(θ = θ),
        iterations = 1,
        data = (y = ybatch, x = xbatch,),
        returnvars = (v = KeepLast(),),
        meta = meta_gp_regression(Xu,kernel_gp),
        free_energy = true,
    )
    return (infer_result.free_energy[end], infer_result.posteriors[:v])
end

Kuu_inverse = (θ) -> cholinv(kernelmatrix(kernel_gp(θ),Xu) .+ 1e-8)

Ψ_0 = (θ;xbatch) -> diag(kernelmatrix(kernel_gp(θ), xbatch))

Ψ_1 = (θ; xbatch) -> begin
    K = kernelmatrix(kernel_gp(θ), xbatch, Xu)
    return K[1:batch_size, :]
end

Ψ_2 = (θ; xbatch) -> begin
    K_Xu = kernelmatrix(kernel_gp(θ), Xu, xbatch)  # Compute kernel between Xu and all points in xtrain_minibatches[1]
    K_xi = kernelmatrix(kernel_gp(θ), xbatch)  # Compute kernel between all points

    # Use a comprehension or matrix operation to compute the product
    return [K_Xu[:, i] * K_Xu[:, i]' for i in 1:batch_size]  # Efficient calculation
end


#for 1 batch
function PerformInference(θ;epochs = 1)
    FE_value = []
    q_v = []
    grad = similar(θ)
    optimizer = Flux.AdaMax()
    θ_optimal = θ
    for epoch=1:epochs
        for b=1:nbatches
            #step 1: Perform inference for v, w 
            fe, qv= my_free_energy(θ_optimal; xbatch = xtrain_minibatches[b], ybatch = ytrain_minibatches[b])
            #step 2: optimize the hyperparameters 
            μ_v, Σ_v = mean_cov(qv)
            Rv = Σ_v + μ_v * μ_v'

            neg_mllh = (θ) -> begin
            # Precompute values that don't change inside the loop
            Ψ_1_values = Ψ_1(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Ψ_0_values = Ψ_0(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Ψ_2_values = Ψ_2(θ; xbatch = xtrain_minibatches[b])  # Compute once
            Kuu_inv = Kuu_inverse(θ)  # Compute once
            Rv_minus_Kuu_inv = Rv - Kuu_inv  # Precompute matrix subtraction
            # Sum over the batch size
            return -sum([
                        1e-3 * ytrain_minibatches[b][i] * (Ψ_1_values[i] * μ_v)[1] - 
                        0.5 * 1e-3 * (Ψ_0_values[i] - tr(Ψ_2_values[i] * Rv_minus_Kuu_inv))
                        for i = 1:batch_size
                    ])
            end

            ForwardDiff.gradient!(grad,neg_mllh, θ_optimal)
            Flux.Optimise.update!(optimizer,θ_optimal,grad)
            append!(FE_value,fe)
            push!(q_v, qv)
        end
    end
    return q_v, FE_value, θ_optimal
end
```

## Inference

```julia
qv, FE, θ_opt = PerformInference(θ_init; epochs = 1);
```
