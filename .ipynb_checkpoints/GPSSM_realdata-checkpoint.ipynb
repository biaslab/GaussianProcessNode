```julia
function compute_derivative(y::Vector, dt::Real)
    n = length(y)
    dy = zeros(n)
    for i=1:n 
        if i==1
            dy[i] = (y[i+1] - y[i]) / dt
        elseif i==n
            dy[i] = y[i] / dt
        else
            dy[i] = (y[i+1] - y[i-1]) / (2*dt)
        end
    end

    return dy 
end
dt = 1/750;
first_derivative_y = compute_derivative(y_data,dt);
second_derivative_y = compute_derivative(first_derivative_y,dt);
y_data_vectorized = [[y_data[i], first_derivative_y[i], second_derivative_y[i]] for i=1:length(y_data)];
```

```julia
# function ReactiveMP.approximate_kernel_expectation(method::AbstractApproximationMethod, g::Function, m::Real, P::Real)

#     weights = ReactiveMP.getweights(method, m, P)
#     points  = ReactiveMP.getpoints(method, m, P)

#     gbar = g(m) - g(m)
#     foreach(zip(weights, points)) do (weight, point)
#         gbar += weight * g(point)
#     end

#     return gbar
# end

# function ReactiveMP.prod(::ProdAnalytical, left::UnivariateGaussianDistributionsFamily, right::ContinuousUnivariateLogPdf) 
#     m,v = approximate_meancov(ghcubature(121),(x) -> exp(right.logpdf(x)),left)
#     return NormalMeanVariance(m,v)
# end

# @rule MvNormalWeightedMeanPrecision(:out, Marginalisation) (q_ξ::PointMass, q_Λ::PointMass, ) = begin 
#     return MvNormalWeightedMeanPrecision(mean(q_ξ),mean(q_Λ))
# end

# ## function for predicting the process
# function predict_gp(X_test, Xu, Yu, kernel)
#     Kuu = kernelmatrix(kernel,Xu)
#     K_test_u = kernelmatrix(kernel,X_test,Xu)
#     Ktest = kernelmatrix(kernel,X_test)

#     μ = K_test_u * inv(Kuu) * Yu 
#     Σ = Ktest - K_test_u * inv(Kuu) * K_test_u'
#     return μ, Σ
# end

struct GPCache
    cache_matrices::Dict{Tuple{Symbol, Tuple{Int, Int}}, Matrix{Float64}}
    cache_vectors::Dict{Tuple{Symbol, Int}, Vector{Float64}}
end

GPCache() = GPCache(Dict{Tuple{Symbol, Tuple{Int, Int}}, Matrix{Float64}}(), Dict{Tuple{Symbol, Int}, Vector{Float64}}())

function getcache(cache::GPCache, label::Tuple{Symbol, Tuple{Int, Int}})
    return get!(() -> Matrix{Float64}(undef, label[2]), cache.cache_matrices, label)
end

function getcache(cache::GPCache, label::Tuple{Symbol, Int})
    return get!(() -> Vector{Float64}(undef, label[2]), cache.cache_vectors, label)
end

function mul_A_B!(cache::GPCache, A::Matrix, B, sizeA1::Int, sizeB2::Int)
    AB = getcache(cache, (:ABdiff, (sizeA1,sizeB2)))
    return mul!(AB, A, B)
end

function mul_A_B!(cache::GPCache, A::Matrix, B, size1::Int)
    #multiply 2 matrices with the same size
    AB = getcache(cache, (:AB, (size1, size1)))
    return mul!(AB, A, B)
end

function mul_A_B_A!(cache::GPCache, A::Matrix, B::Matrix, size1::Int)
    #A, B are square matrices with a same size
    AB = getcache(cache, (:AB, (size1, size1)))
    ABA = getcache(cache, (:ABA, (size1, size1)))
    mul!(AB, A, B)

    return mul!(ABA, AB, A)
end

function mul_A_B_At!(cache::GPCache, A::Matrix, B::Matrix, sizeA1::Int, sizeB1::Int)
    # A: matrix with size (sizeA1,sizeB1)
    # B: matrix with size (sizeB1,sizeB1)
    AB = getcache(cache, (:AB, (sizeA1, sizeB1)))
    ABAt = getcache(cache, (:ABA, (sizeA1, sizeA1)))
    mul!(AB, A, B)
    return mul!(ABAt, AB, A')
end
function mul_A_v!(cache::GPCache, A::Matrix, v::Vector, sizeA1::Int)
    Av = getcache(cache, (:Av, sizeA1))
    return mul!(Av, A, v)
end

# ## GP meta 
struct GPTransitionMeta{I,T,K}
    Xu          :: I    # inducing inputs
    Kuu_inverse :: T  
    C           :: Union{Nothing,T}    # co-regionalization matrix 
    kernel      :: K
    GPCache     :: Union{Nothing,GPCache}
end
getInducingInput(meta::GPTransitionMeta) = meta.Xu
getinverseKuu(meta::GPTransitionMeta) = meta.Kuu_inverse
getCoregionalizationMatrix(meta::GPTransitionMeta) = meta.C
getKernel(meta::GPTransitionMeta) = meta.kernel
getGPCache(meta::GPTransitionMeta) = meta.GPCache

# ## specify GP node 
struct GPTransitionControl end 

@node GPTransitionControl Stochastic [ out, in, u , w, c] ## out: x_t , in: x_{t-1}, c:control signal,  u: inducing points , w: precision of process noise 

# ## specify rules
# @rule GPTransitionControl(:out, Marginalisation) (q_in::UnivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::GammaShapeRate, q_c::PointMass, meta::GPTransitionMeta,) = begin
#     Kuu_inverse = getinverseKuu(meta)
#     Xu = getInducingInput(meta)
#     kernel = getKernel(meta)
#     c = mean(q_c)
#     μ_u = mean(q_u)
#     kxu = (x) -> kernelmatrix(kernel, [[x,c]], Xu) 
#     Ψ1 = ReactiveMP.approximate_kernel_expectation(ghcubature(121), kxu, q_in)
#     return NormalMeanPrecision((Ψ1 * Kuu_inverse * μ_u)[1], mean(q_w))
# end

# @rule GPTransitionControl(:in, Marginalisation) (q_out::UnivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::GammaShapeRate, q_c::PointMass, meta::GPTransitionMeta) = begin
#     w_bar = mean(q_w) # this is just temporary assumption, will need to add this interface for the node 
#     c = mean(q_c)

#     E_logw = mean(log,q_w)
#     kernel = getKernel(meta)
#     Xu = getInducingInput(meta)
#     Kuu_inverse = getinverseKuu(meta)
#     A = (x) -> kernelmatrix(kernel,[[x,c]],[[x,c]]) .- kernelmatrix(kernel,[[x,c]],Xu) * Kuu_inverse * kernelmatrix(kernel,Xu,[[x,c]])
#     B = (x) -> kernelmatrix(kernel, [[x,c]], Xu) * Kuu_inverse
    
#     μ_y, v_y = mean_var(q_out)
#     μ_u, Σ_u = mean_cov(q_u)

#     log_backwardmess = (x) -> -0.5 * log(2π) + 0.5*E_logw - 0.5 * w_bar * (A(x)[1] + v_y + tr(B(x)' * B(x) *Σ_u) + (μ_y - (B(x)*μ_u)[1])^2)
#     return ContinuousUnivariateLogPdf(log_backwardmess)
# end

# @rule GPTransitionControl(:u, Marginalisation) (q_out::UnivariateNormalDistributionsFamily, q_in::UnivariateNormalDistributionsFamily, q_w::GammaShapeRate,q_c::PointMass, meta::GPTransitionMeta) = begin
#     w = mean(q_w) # this is just temporary assumption, will need to add this interface for the node 
#     c = mean(q_c)
#     μ_y = mean(q_out)
#     Xu = getInducingInput(meta)
#     Kuu_inverse = getinverseKuu(meta)
#     kernel = getKernel(meta)

#     Ψ1 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, [[x,c]], Xu),q_in)
#     Ψ2 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, Xu, [[x,c]]) * kernelmatrix(kernel, [[x,c]], Xu), q_in) + 1e-7*I 

#     ξ_u = μ_y * w * Kuu_inverse * Ψ1' #weighted-mean
#     W_u = w * Kuu_inverse * Ψ2 * Kuu_inverse #precision 
#     return MvNormalWeightedMeanPrecision(vcat(ξ_u...), W_u)
# end

# @rule GPTransitionControl(:w, Marginalisation) (q_out::UnivariateGaussianDistributionsFamily, q_in::UnivariateGaussianDistributionsFamily,q_u::MultivariateNormalDistributionsFamily, q_c::PointMass, meta::GPTransitionMeta,) = begin
#     μ_y, v_y = mean_var(q_out)
#     μ_u, Σ_u = mean_cov(q_u)
#     c = mean(q_c)
#     Xu = getInducingInput(meta)
#     Kuu_inverse = getinverseKuu(meta)
#     kernel = getKernel(meta)
#     Ψ0 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, [[x,c]], [[x,c]]),q_in)[]
#     Ψ1 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, [[x,c]], Xu),q_in)
#     Ψ2 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, Xu, [[x,c]]) * kernelmatrix(kernel, [[x,c]], Xu), q_in) + 1e-7*I 
    
#     I1 = clamp(Ψ0 - tr(Kuu_inverse * Ψ2),1e-12,1e12)
#     I2 = clamp(μ_y^2 + v_y -2*μ_y * (Ψ1 * Kuu_inverse * μ_u)[] + tr(Kuu_inverse * (Σ_u + μ_u * μ_u') * Kuu_inverse * Ψ2 ), 1e-12,1e12)
#     return GammaShapeRate(1.5, 0.5*(I1 + I2))
# end

# @average_energy GPTransitionControl (q_out::UnivariateNormalDistributionsFamily, q_in::UnivariateGaussianDistributionsFamily, q_u::MultivariateNormalDistributionsFamily, q_w::GammaShapeRate, q_c::PointMass, meta::GPTransitionMeta,) = begin
#     w_bar = mean(q_w)
#     c = mean(q_c)
#     E_logw = mean(log,q_w)
#     μ_y, v_y = mean_var(q_out)
#     μ_u, Σ_u = mean_cov(q_u)
#     Xu = getInducingInput(meta)
#     Kuu_inverse = getinverseKuu(meta)
#     kernel = getKernel(meta)
#     Ψ0 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, [[x,c]], [[x,c]]),q_in)[]
#     Ψ1 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, [[x,c]], Xu),q_in)
#     Ψ2 = ReactiveMP.approximate_kernel_expectation(ghcubature(121),(x) -> kernelmatrix(kernel, Xu, [[x,c]]) * kernelmatrix(kernel, [[x,c]], Xu), q_in) + 1e-7*I 

#     I1 = clamp(Ψ0 - tr(Kuu_inverse * Ψ2),1e-12,1e12)
#     I2 = clamp(μ_y^2 + v_y -2*μ_y * (Ψ1 * Kuu_inverse * μ_u)[] + tr(Kuu_inverse * (Σ_u + μ_u * μ_u') * Kuu_inverse * Ψ2 ), 1e-12,1e12)

#     return 0.5*(I1*w_bar - E_logw + log(2π) + I2 * w_bar) 
# end
```

### Multivariate case

```julia
## Multivariate case 
kernel_value_Ψ0 = (x,control, kernel) -> getindex(kernelmatrix(kernel,[[x,control]]),1)
compute_hessian_Ψ0 = (x, control, kernel) -> Zygote.hessian((z) -> kernel_value_Ψ0(z,control,kernel),x)
function approx_Ψ0_Laplace(q_x::T, control, kernel) where {T <: Distribution}
    x0, Σ = mean_cov(q_x)
    hessian_term = tr(compute_hessian_Ψ0(x0, control, kernel) * Σ)
    constant_terms = kernel_value_Ψ0(x0, control, kernel)
    return constant_terms + 0.5*hessian_term
end

kernel_value = (x,control,y, kernel) -> getindex(kernelmatrix(kernel,[[x,control]],[y]),1)
compute_hessian_Ψ1 = (x,control,y, kernel) -> ForwardDiff.hessian((z) -> kernel_value(z,control,y,kernel),x)
hessian_vector = (x, control, Xu, kernel) -> [compute_hessian_Ψ1(x,control,xu,kernel) for xu in Xu]
function approx_Ψ1_Laplace!(cache::GPCache, Ψ1::Vector, Xu::Vector,q_x::T, control, kernel) where {T <: Distribution}
    x0, Σ = mean_cov(q_x)
    Ψ1 = 0.5*[tr(mul_A_B!(cache,A,Σ,length(x0))) for A in hessian_vector(x0,control,Xu,kernel)]
    Ψ1 += [kernel_value(x0,control,xu,kernel) for xu in Xu]
    return Ψ1
end

cov_kernel_val = (x,control,y,z,kernel) -> getindex(kernelmatrix(kernel,[[x,control]],[y]),1) * getindex(kernelmatrix(kernel,[[x,control]],[z]),1)
compute_hessian_Ψ2 = (x,control,y,z,kernel) -> ForwardDiff.hessian((v) -> cov_kernel_val(v,control,y,z,kernel), x)

function approximate_Ψ2_Laplace!(cache::GPCache, Ψ2::Matrix, Xu::Vector,q_x::T, control, kernel) where {T <: Distribution}
    x0, Σ = mean_cov(q_x)
    m = length(Xu)
    # Ψ2 = zeros(m,m)
    Ψ2_0 = kernelmatrix(kernel,Xu,[x0,control]) * kernelmatrix(kernel,[x0,control],Xu)
    for j=1:m
        for i=1:m
            Ψ2[i,j] = Ψ2_0[i,j] + 0.5 * tr(mul_A_B!(cache, compute_hessian_Ψ2(x0,control,Xu[i],Xu[j],kernel), Σ, length(x0)))
        end
    end
    return Ψ2 
end

Afunc = (x,control,kernel) -> kron(C,kernelmatrix(kernel,[[x,control]],[[x,control]])[] - getindex(kernelmatrix(kernel,[[x,control]],Xu) * Kuu_inverse * kernelmatrix(kernel,Xu,[[x,control]]),1))
Bfunc = (x,control,kernel) -> kron(C,kernelmatrix(kernel, [[x,control]], Xu))
```

```julia
#out rule 
@rule GPTransitionControl(:out, Marginalisation) (q_in::MultivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::PointMass,q_c::PointMass, meta::GPTransitionMeta,) = begin
    μ_u = mean(q_u)
    W = mean(q_w) 
    control_signal = mean(q_c)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)

    M = length(Xu) #number of inducing points
    D = length(mean(q_in)) #dimension 
    
    @inbounds μ_u = [view(μ_u,i:i+M-1) for i=1:M:M*D] 
    Ψ1 = getcache(getGPCache(meta),(:Ψ1, M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1, Xu,q_in, control_signal, kernel)
    μ_y = getcache(getGPCache(meta), (:μ_y, D))
    μ_u_transformed = getcache(getGPCache(meta), (:μ_u_transformed, M))
    for (i,μ_u_entry) in enumerate(μ_u)
        mul!(μ_u_transformed,Kuu_inverse,μ_u_entry)
        @inbounds μ_y[i] = jdotavx(Ψ1,μ_u_transformed)
    end
    return MvNormalMeanPrecision(μ_y, W)
end

@rule GPTransitionControl(:out, Marginalisation) (q_in::MultivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::Wishart, q_c::PointMass, meta::GPTransitionMeta,) = begin
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(mean(q_in)) #dimension 
    μ_u = mean(q_u)
    W = mean(q_w) 
    control_signal = mean(q_c)

    @inbounds μ_u = [view(μ_u,i:i+M-1) for i=1:M:M*D] 
    Ψ1 = getcache(getGPCache(meta),(:Ψ1, M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1, Xu,q_in,control_signal,kernel)
    μ_y = getcache(getGPCache(meta), (:μ_y, D))
    μ_u_transformed = getcache(getGPCache(meta), (:μ_u_transformed, M))
    for (i,μ_u_entry) in enumerate(μ_u)
        mul!(μ_u_transformed,Kuu_inverse,μ_u_entry)
        @inbounds μ_y[i] = jdotavx(Ψ1,μ_u_transformed)
    end
    return MvNormalMeanPrecision(μ_y, W)
end

#in rule 
@rule GPTransitionControl(:in, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::PointMass,q_c::PointMass, meta::GPTransitionMeta) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_u, Σ_u = mean_cov(q_u)
    W = mean(q_w) 
    control_signal = mean(q_c)
    E_logW = log(det(W))
    Xu = getInducingInput(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    
    neg_log_backwardmess = (x) -> -(-0.5*D * log(2π) + 0.5*E_logW - 0.5  * (tr(W*(Afunc(x,control_signal,kernel) + Σ_y + Bfunc(x,control_signal,kernel)*Σ_u*Bfunc(x,control_signal,kernel)')) + (μ_y - Bfunc(x,control_signal,kernel)*μ_u)' * W * (μ_y - Bfunc(x,control_signal,kernel)*μ_u)))
    res = optimize(neg_log_backwardmess,zeros(D))
    m_z = res.minimizer
    W_z = Zygote.hessian(neg_log_backwardmess, m_z) 
    return MvNormalWeightedMeanPrecision(W_z * m_z, W_z)
end

@rule GPTransitionControl(:in, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily,q_w::Wishart, q_c::PointMass, meta::GPTransitionMeta) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_u, Σ_u = mean_cov(q_u)
    control_signal = mean(q_c)
    W = mean(q_w) 
    E_logW = mean(logdet,q_w)
    Xu = getInducingInput(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    
    neg_log_backwardmess = (x) -> -(-0.5*D * log(2π) + 0.5*E_logW - 0.5  * (tr(W*(Afunc(x,control_signal,kernel) + Σ_y + Bfunc(x,control_signal,kernel)*Σ_u*Bfunc(x,control_signal,kernel)')) + (μ_y - Bfunc(x,control_signal,kernel)*μ_u)' * W * (μ_y - Bfunc(x,control_signal,kernel)*μ_u)))
    res = optimize(neg_log_backwardmess,zeros(D))
    m_z = res.minimizer
    W_z = Zygote.hessian(neg_log_backwardmess, m_z) 
    return MvNormalWeightedMeanPrecision(W_z * m_z, W_z)
end

# u rule
@rule GPTransitionControl(:u, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateNormalDistributionsFamily, q_w::PointMass, q_c::PointMass, meta::GPTransitionMeta) = begin
    W = mean(q_w)
    μ_y = mean(q_out)
    control_signal = mean(q_c)
    Kuu_inverse = getinverseKuu(meta)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    
    #Remember to change Unit_Normal
    Ψ1 = getcache(cache,(:Ψ1,M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1,Xu,q_in,control_signal,kernel)
    Ψ2 = approximate_Ψ2_Laplace(gpcache,Xu, q_in, control_signal,kernel)

    #we'll use NormalWeightedMeanPrecision
    ξ_u = Float64[]
    Ψ1_transformed = mul_A_B!(cache,Kuu_inverse,Ψ1,M,1) #Ψ1_transformed = Kuu_inverse * Ψ1
    μ_y_transformed = W * μ_y 
    for mu_y in μ_y_transformed
        ξ = mu_y .* Ψ1_transformed
        append!(ξ_u, ξ)
    end

    Ψ2_transformed = mul_A_B_A!(cache, Kuu_inverse, Ψ2,M) # Ψ2_transformed = Kuu_inverse * Ψ2 * Kuu_inverse
    W_u = getcache(cache, (:W_u, (D*M,D*M)))
    kron!(W_u,W, Ψ2_transformed)
    return MvNormalWeightedMeanPrecision(ξ_u, W_u)
end

# w rule 
@rule GPTransitionControl(:w, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateNormalDistributionsFamily, q_u::MultivariateNormalDistributionsFamily, q_c::PointMass, meta::GPTransitionMeta,) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_u, Σ_u = mean_cov(q_u)
    control_signal = mean(q_c)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)

    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    
    μ_u = [view(μ_u,i:i+M-1) for i=1:M:M*D]  
    Σ_u = create_blockmatrix(Σ_u,D,M)

    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1,Xu,q_in,control_signal,kernel)
    Ψ2 = approximate_Ψ2_Laplace(gpcache,Xu, q_in,control_signal,kernel)

    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (approx_Ψ0_Laplace(q_in,control_signal,kernel) - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    B_mat = mul_A_B!(cache,Kuu_inverse,Ψ1,M,1) #B = Ψ1 * Kuu_inverse
    Ψ2_transformed = mul_A_B_A!(cache, Kuu_inverse, Ψ2,M) # Kuu_inverse * Ψ2 * Kuu_inverse
    E = [getindex(B_mat' * i,1) for i in μ_u]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_u[i]*μ_u[j]' + Σ_u[i,j]) * Ψ2_transformed)
        end
    end
    I2 = μ_y * μ_y' + Σ_y - μ_y * E' - E * μ_y' + Ψ_5 
    return WishartMessage(D+2, I1 + I2)
end

## average energy
@average_energy GPTransitionControl (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateGaussianDistributionsFamily, q_u::MultivariateNormalDistributionsFamily, q_w::PointMass, q_c::PointMass, meta::GPTransitionMeta,) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_u, Σ_u = mean_cov(q_u)
    control_signal = mean(q_c)
    W = mean(q_w)
    E_logW = log(det(W))
    C = getCoregionalizationMatrix(meta)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension

    μ_u = [view(μ_u,i:i+M-1) for i=1:M:M*D]  
    Σ_u = create_blockmatrix(Σ_u,D,M)

    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1,Xu,q_in,control_signal,kernel)
    Ψ2 = approximate_Ψ2_Laplace(gpcache,Xu, q_in, control_signal,kernel)

    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (approx_Ψ0_Laplace(q_in, control_signal,kernel) - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    B_mat = mul_A_B!(cache,Kuu_inverse,Ψ1,M,1) #B = Ψ1 * Kuu_inverse
    Ψ2_transformed = mul_A_B_A!(cache, Kuu_inverse, Ψ2,M) # Kuu_inverse * Ψ2 * Kuu_inverse
    E = [getindex(B_mat' * i,1) for i in μ_u]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_u[i]*μ_u[j]' + Σ_u[i,j]) * Ψ2_transformed)
        end
    end
    I2 = μ_y * μ_y' + Σ_y - μ_y * E' - E * μ_y' + Ψ_5 

    return 0.5*tr(W*I1) + 0.5*D*log(2π) - 0.5*E_logW + 0.5 * tr(W * I2)
end

@average_energy GPTransitionControl (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateGaussianDistributionsFamily, q_u::MultivariateNormalDistributionsFamily, q_w::PointMass, q_c::PointMass, meta::GPTransitionMeta,) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_u, Σ_u = mean_cov(q_u)
    control_signal = mean(q_c)
    W = mean(q_w)
    E_logW = mean(logdet,q_w)
    C = getCoregionalizationMatrix(meta)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension

    μ_u = [view(μ_u,i:i+M-1) for i=1:M:M*D]  
    Σ_u = create_blockmatrix(Σ_u,D,M)

    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ1 = approx_Ψ1_Laplace!(Ψ1,Xu,q_in,control_signal,kernel)
    Ψ2 = approximate_Ψ2_Laplace(gpcache,Xu, q_in, control_signal,kernel)

    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (approx_Ψ0_Laplace(q_in, control_signal,kernel) - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    B_mat = mul_A_B!(cache,Kuu_inverse,Ψ1,M,1) #B = Ψ1 * Kuu_inverse
    Ψ2_transformed = mul_A_B_A!(cache, Kuu_inverse, Ψ2,M) # Kuu_inverse * Ψ2 * Kuu_inverse
    E = [getindex(B_mat' * i,1) for i in μ_u]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_u[i]*μ_u[j]' + Σ_u[i,j]) * Ψ2_transformed)
        end
    end
    I2 = μ_y * μ_y' + Σ_y - μ_y * E' - E * μ_y' + Ψ_5 

    return 0.5*tr(W*I1) + 0.5*D*log(2π) - 0.5*E_logW + 0.5 * tr(W * I2)
end
```

## Create Model

```julia
@model function multisine_model(n, Kuu_inverse, control_signal)
    y = datavar(Vector{Float64},n)
    x = randomvar(n)

    u ~ MvNormalWeightedMeanPrecision(zeros(size(Kuu_inverse,1)), Kuu_inverse)
    # w ~ Wishart(1e3,diageye(3))
    x0 ~ MvNormalMeanCovariance(zeros(3), 50 * diageye(3))
    x_prev = x0
    for i=1:n
        x[i] ~ GPTransitionControl(x_prev, u, 0.1*diageye(3), control_signal[i]) where { q = MeanField() }  #mean-field assumption
        y[i] ~ MvNormalMeanCovariance(x[i],[0.0001 0 0; 0 0.1 0; 0 0 1.])
        x_prev = x[i]
    end
end
```

```julia
function UpdateState(prior_prediction, observation)
    likelihood_update = @call_rule NormalMeanVariance(:μ, Marginalisation) (m_out = PointMass(observation), m_v = PointMass(0.001))
    x = ReactiveMP.prod(ProdAnalytical(), prior_prediction, likelihood_update)
    return x
end

function  PredictState(prev_state, control_signal)
    return @call_rule GPTransitionControl(:out, Marginalisation) (q_in = prev_state, q_u = qu, q_w = qw, q_c = PointMass(control_signal),meta = GPTransitionMeta(Xu,Kuu_inverse,nothing,kernel_gp,nothing))
end

function  PredictObservation(prev_state, control_signal)
    q_state = PredictState(prev_state, control_signal)
    return @call_rule NormalMeanVariance(:out, Marginalisation) (m_μ = q_state, m_v = PointMass(0.001))
end

#perform 1-step-ahead update 
prev_state = qx[end]
predicted_state = []
predicted_observation = []
for i=1:n_trial
    prediction = PredictState(prev_state, test_control_signal[i])
    predictive_obser = PredictObservation(prev_state, test_control_signal[i])
    push!(predicted_state,prediction)
    push!(predicted_observation, predictive_obser)
    prev_state = UpdateState(prediction, test_data[i])
    # prev_state = prediction
end
;
```
