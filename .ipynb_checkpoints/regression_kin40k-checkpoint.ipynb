## Load data

```julia
# load data 
xtrain_data = matopen("data/kin40k/kin40k_xtrain.mat");
xtest_data = matopen("data/kin40k/kin40k_xtest.mat");
ytrain_data = matopen("data/kin40k/kin40k_ytrain.mat");
ytest_data = matopen("data/kin40k/kin40k_ytest.mat");

xtrain_matrix = read(xtrain_data, "xtrain");
ytrain = read(ytrain_data, "ytrain") |> (x) -> vcat(x...);
xtest_matrix = read(xtest_data, "xtest");
ytest = read(ytest_data, "ytest") |> (x) -> vcat(x...);

Ntrain = length(ytrain);
xtrain = [xtrain_matrix[i,:] for i=1:Ntrain];
data_training = (xtrain, ytrain);

Ntest = length(ytest);
xtest = [xtest_matrix[i,:] for i=1:Ntest];

batch_size = 4;
xtrain_minibatches, ytrain_minibatches = split2batch(data_training, batch_size);
nbatches = length(ytrain_minibatches);
```

## Configure GP

```julia
# configure GP 
Random.seed!(1)
M = 600;

pos = randperm(Ntrain)[1:M]
Xu = xtrain[pos]; #inducing input 

kernel_gp(θ) = (StatsFuns.softplus(θ[1])) * with_lengthscale(SEKernel(),StatsFuns.softplus.(θ[2:end]))

dim_θ = size(xtrain_matrix,2) + 1
gpcache = GPCache();
θ_init = StatsFuns.invsoftplus.(ones(dim_θ));
optimizer = Flux.AdaMax();

Ψ0 = [1.0;;]
Ψ1_trans = kernelmatrix(kernel_gp(θ_init),Xu,[xtrain[1]])
Ψ2 = kernelmatrix(kernel_gp(θ_init),Xu,[xtrain[1]]) * kernelmatrix(kernel_gp(θ_init),[xtrain[1]],Xu);
w_val = 0.01
```

## Model Definition

```julia
@model function gp_kin40k(y,x,θ, mv, Σv, shape, rate)
    v ~ MvNormalMeanCovariance(mv, Σv)
    w ~ GammaShapeRate(shape, rate)
    for i in eachindex(y)
        y[i] ~ UniSGP(x[i],v,w,θ) 
    end
end
gp_constraints = @constraints begin
    q(v,w) = q(v)q(w)
end

gp_initialization = @initialization begin
    q(w) = GammaShapeRate(1e-2,1e-2)
end

@meta function meta_gp_regression(Xu,Ψ0,Ψ1_trans,Ψ2,Kuu_inverse,kernel, cache)
    UniSGP() -> UniSGPMeta(nothing,Xu,Ψ0,Ψ1_trans,Ψ2,Kuu_inverse,kernel,cache)
end
```

## Inference

```julia
function my_free_energy(θ; xbatch, ybatch, mv, Σv,shape, rate,niterations)
    Kuu = kernelmatrix(kernel_gp(θ), Xu) + 1e-8 * I
    KuuL = fastcholesky!(Kuu).L
    # gpcache.cache_LowerTriangular[(:KuuL,M)] = KuuL
    infer_result = infer(
        model = gp_kin40k(θ = θ, mv = mv, Σv = Σv,shape = shape, rate = rate,),
        iterations = niterations,
        data = (y = ybatch, x = xbatch,),
        initialization = gp_initialization,
        constraints = gp_constraints,
#         options = (limit_stack_depth = 400, ), 
        returnvars = (v = KeepLast(),w = KeepLast(),),
        meta = meta_gp_regression(Xu,Ψ0,Ψ1_trans,Ψ2,KuuL,kernel_gp,gpcache),
        free_energy = false,
    )
    # return (infer_result.free_energy[end], infer_result.posteriors[:v], infer_result.posteriors[:w])
    return (infer_result.posteriors[:v], infer_result.posteriors[:w])
end
function PerformInference(θ;epochs = 1)
    FE_value = []
    μ_v = zeros(M)
    Σ_v = 50*diageye(M)
    shape = 0.01
    rate = 0.01
    w = 0.
    grad = similar(θ)
    θ_optimal = θ

    @inbounds for epoch=1:epochs
        @inbounds for b=1:1
            #step 1: Perform inference for v, w 
            qv, qw = my_free_energy(θ_optimal;   xbatch = xtrain_minibatches[b], 
                                                ybatch = ytrain_minibatches[b],
                                                mv = μ_v,
                                                Σv = Σ_v,
                                                shape=shape,
                                                rate = rate,
                                                niterations=1)
            #step 2: optimize the hyperparameters 
            μ_v, Σ_v = mean_cov(qv)
            shape,rate = params(qw)
            w = mean(qw)
        end
#         Rv = Σ_v + μ_v * μ_v' |> (x) -> fastcholesky!(x).L
#         for b=1:nbatches
#             grad_llh_new!(grad,θ_optimal; y_data=ytrain_minibatches[b],
#                                     x_data=xtrain_minibatches[b],
#                                     v = μ_v,
#                                     Lv=Rv,
#                                     w=w,
#                                     kernel=kernel_gp,
#                                     Xu=Xu)
#             Flux.Optimise.update!(optimizer,θ_optimal,grad)
#         end
            # append!(FE_value,fe)
    end
    q_v = MvNormalMeanCovariance(μ_v, Σ_v)
    return FE_value, q_v,θ_optimal
end
```

```julia
@time FE, qv, θ_opt = PerformInference(θ_init; epochs = 1);
```

```julia
# predict function 
function predict_new(x_test,qv,qw,θ_opt, meta)
    prediction = @call_rule UniSGP(:out, Marginalisation) ( q_in=PointMass(x_test),
                                                            q_v = qv, 
                                                            q_w = qw,
                                                            q_θ=PointMass(θ_opt), 
                                                            meta=meta)
    return prediction
end
Kuu = kernelmatrix(kernel_gp(θ_opt), Xu) + 1e-8 * I
KuuL = fastcholesky!(Kuu).L
predict_mean =[]
for i=1:Ntest
    prediction = predict_new(xtest[i],qv,PointMass(0.1),θ_opt,UniSGPMeta(nothing,Xu,Ψ0,Ψ1_trans,Ψ2,KuuL,kernel_gp,GPCache()))
    append!(predict_mean,mean(prediction))
end
```
