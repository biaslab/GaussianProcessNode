```julia
function approximate_kernel_expectation(method::AbstractApproximationMethod, g::Function, m::AbstractVector{T}, P::AbstractMatrix{T}) where {T <: Real}
    weights = ReactiveMP.getweights(method, m, P)
    points  = ReactiveMP.getpoints(method, m, P)

    gbar = g(m) - g(m)
    foreach(zip(weights, points)) do (weight, point)
        axpy!(weight, g(point), gbar) # gbar = gbar + weight * g(point)
    end

    return gbar
end
```

## Define GP

```julia
struct GPTransition end 

@node GPTransition Stochastic [ out, in, v , w] ## out: x_t , in: x_{t-1},  v: transformed-inducing points Kuu_inv * u , w: precision of process noise 
```

```julia
@rule GPTransition(:out, Marginalisation) (q_in::MultivariateNormalDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::Wishart,meta::GPTransitionMeta,) = begin
    μ_v = mean(q_v)
    W = mean(q_w) 
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    C = getCoregionalizationMatrix(meta)
    M = length(Xu) #number of inducing points
    D = size(C,1)
    @inbounds μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D] 
    μ_y = getcache(getGPCache(meta), (:μ_y, D))
    Ψ1 = approximate_kernel_expectation(srcubature(), (x) -> kernelmatrix(kernel, [x], Xu), q_in)
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    for (i,μ_v_entry) in enumerate(μ_v)
        @inbounds μ_y[i] = jdotavx(Ψ1,μ_v_entry)
    end
    return MvNormalMeanPrecision(μ_y, W)
end

@rule GPTransition(:out, Marginalisation) (q_in::MultivariateNormalDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::PointMass,meta::GPTransitionMeta,) = begin
    μ_v = mean(q_v)
    W = mean(q_w) 
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    C = getCoregionalizationMatrix(meta)
    M = length(Xu) #number of inducing points
    D = size(C,1)
    @inbounds μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D] 
    μ_y = getcache(getGPCache(meta), (:μ_y, D))
    Ψ1 = approximate_kernel_expectation(srcubature(), (x) -> kernelmatrix(kernel, [x], Xu), q_in)
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    for (i,μ_v_entry) in enumerate(μ_v)
        @inbounds μ_y[i] = jdotavx(Ψ1,μ_v_entry)
    end
    return MvNormalMeanPrecision(μ_y, W)
end
#in rule 
@rule GPTransition(:in, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_v::MultivariateNormalDistributionsFamily,q_w::Wishart, meta::GPTransitionMeta) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W = mean(q_w) 
    kernel = getKernel(meta)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    D = length(μ_y) #output dimension

    kernel = getKernel(meta)
    A = (x) -> kron(C,kernelmatrix(kernel,[x]) - kernelmatrix(kernel,[x],Xu)*Kuu_inverse*kernelmatrix(kernel,Xu,[x]))
    B = (x) -> kron(C,kernelmatrix(kernel, [x], Xu))
    
    neg_log_backwardmess = (x) -> 0.5  * (tr(W*(A(x) + Σ_y + B(x)*Σ_v*B(x)')) + (μ_y - B(x)*μ_v)' * W * (μ_y - B(x)*μ_v))
    res = optimize(neg_log_backwardmess,zeros(length(Xu[1])))
    m_z = res.minimizer
    W_z = Zygote.hessian(neg_log_backwardmess, m_z) 
 
    return MvNormalWeightedMeanPrecision(W_z * m_z, W_z)
end

@rule GPTransition(:in, Marginalisation) (q_out::PointMass, q_v::MultivariateNormalDistributionsFamily,q_w::Wishart, meta::GPTransitionMeta) = begin
    μ_y = mean(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W = mean(q_w) 
    kernel = getKernel(meta)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    D = length(μ_y) #output dimension

    kernel = getKernel(meta)
    A = (x) -> kron(C,kernelmatrix(kernel,[x]) - kernelmatrix(kernel,[x],Xu)*Kuu_inverse*kernelmatrix(kernel,Xu,[x]))
    B = (x) -> kron(C,kernelmatrix(kernel, [x], Xu))
    
    neg_log_backwardmess = (x) -> 0.5  * (tr(W*(A(x)+ B(x)*Σ_v*B(x)')) + (μ_y - B(x)*μ_v)' * W * (μ_y - B(x)*μ_v))
    res = optimize(neg_log_backwardmess,zeros(length(Xu[1])))
    m_z = res.minimizer
    W_z = Zygote.hessian(neg_log_backwardmess, m_z) 
 
    return MvNormalWeightedMeanPrecision(W_z * m_z, W_z)
end

@rule GPTransition(:in, Marginalisation) (q_out::PointMass, q_v::MultivariateNormalDistributionsFamily,q_w::PointMass, meta::GPTransitionMeta) = begin
    μ_y = mean(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W = mean(q_w) 
    kernel = getKernel(meta)
    Kuu_inverse = getinverseKuu(meta)
    Xu = getInducingInput(meta)
    D = length(μ_y) #output dimension

    kernel = getKernel(meta)
    A = (x) -> kron(C,kernelmatrix(kernel,[x]) - kernelmatrix(kernel,[x],Xu)*Kuu_inverse*kernelmatrix(kernel,Xu,[x]))
    B = (x) -> kron(C,kernelmatrix(kernel, [x], Xu))
    
    neg_log_backwardmess = (x) -> 0.5  * (tr(W*(A(x)+ B(x)*Σ_v*B(x)')) + (μ_y - B(x)*μ_v)' * W * (μ_y - B(x)*μ_v))
    @btime res = optimize(neg_log_backwardmess,zeros(length(Xu[1])))
    m_z = res.minimizer
    W_z = Zygote.hessian(neg_log_backwardmess, m_z) 
 
    return MvNormalWeightedMeanPrecision(W_z * m_z, W_z)
end

#v rule
@rule GPTransition(:v, Marginalisation) (q_out::MultivariateGaussianDistributionsFamily, q_in::MultivariateGaussianDistributionsFamily, q_w::Wishart, meta::GPTransitionMeta) = begin 
    W = mean(q_w)
    μ_y = mean(q_out)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension

    Ψ1 = getcache(cache,(:Ψ1,M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    #we'll use NormalWeightedMeanPrecision
    ξ_v = Float64[]
    μ_y_transformed = W * μ_y 
    for mu_y in μ_y_transformed
        append!(ξ_v, mu_y .* Ψ1)
    end
    W_v = getcache(cache, (:W_v, (D*M,D*M)))
    kron!(W_v,W, Ψ2)
    return MvNormalWeightedMeanPrecision(ξ_v, W_v)
end

@rule GPTransition(:v, Marginalisation) (q_out::PointMass, q_in::MultivariateGaussianDistributionsFamily, q_w::Wishart, meta::GPTransitionMeta) = begin 
    W = mean(q_w)
    μ_y = mean(q_out)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension

    Ψ1 = getcache(cache,(:Ψ1,M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    #we'll use NormalWeightedMeanPrecision
    ξ_v = Float64[]
    μ_y_transformed = W * μ_y 
    for mu_y in μ_y_transformed
        append!(ξ_v, mu_y .* Ψ1)
    end
    W_v = getcache(cache, (:W_v, (D*M,D*M)))
    kron!(W_v,W, Ψ2)
    return MvNormalWeightedMeanPrecision(ξ_v, W_v)
end

@rule GPTransition(:v, Marginalisation) (q_out::PointMass, q_in::MultivariateGaussianDistributionsFamily, q_w::PointMass, meta::GPTransitionMeta) = begin 
    W = mean(q_w)
    μ_y = mean(q_out)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    Xu = getInducingInput(meta)
    kernel = getKernel(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension

    Ψ1 = getcache(cache,(:Ψ1,M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    #we'll use NormalWeightedMeanPrecision
    ξ_v = Float64[]
    μ_y_transformed = W * μ_y 
    for mu_y in μ_y_transformed
        append!(ξ_v, mu_y .* Ψ1)
    end
    W_v = getcache(cache, (:W_v, (D*M,D*M)))
    kron!(W_v,W, Ψ2)
    return MvNormalWeightedMeanPrecision(ξ_v, W_v)
end

# w rule 
@rule GPTransition(:w, Marginalisation) (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateNormalDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, meta::GPTransitionMeta,) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    Xu = getInducingInput(meta)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D]  
    Σ_v = create_blockmatrix(Σ_v,D,M)
    Kuu_inverse = getinverseKuu(meta)
    kernel = getKernel(meta)

    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ0 = approximate_expectation(GenUnscented(),q_in, (x) -> kernelmatrix(kernel, [x], [x]))[]
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 

    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (Ψ0 - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    E = [getindex(Ψ1 * i,1) for i in μ_v]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_v[i]*μ_v[j]' + Σ_v[i,j]) * Ψ2)
        end
    end
    I2 = μ_y * μ_y' + Σ_y - μ_y * E' - E * μ_y' + Ψ_5 
    return WishartFast(D+2, I1 + I2)
end
@rule GPTransition(:w, Marginalisation) (q_out::PointMass, q_in::MultivariateGaussianDistributionsFamily, q_v::MultivariateGaussianDistributionsFamily, meta::GPTransitionMeta) = begin 
    μ_y = mean(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    Xu = getInducingInput(meta)
    C = getCoregionalizationMatrix(meta)
    cache = getGPCache(meta)
    M = length(Xu) #number of inducing points
    D = length(μ_y) #dimension
    μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D]  
    Σ_v = create_blockmatrix(Σ_v,D,M)
    Kuu_inverse = getinverseKuu(meta)
    kernel = getKernel(meta)

    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ0 = approximate_expectation(GenUnscented(),q_in, (x) -> kernelmatrix(kernel, [x], [x]))[]
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 

    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (Ψ0 - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    E = [getindex(Ψ1 * i,1) for i in μ_v]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_v[i]*μ_v[j]' + Σ_v[i,j]) * Ψ2)
        end
    end
    I2 = μ_y * μ_y' - μ_y * E' - E * μ_y' + Ψ_5 
    return WishartFast(D+2, I1 + I2)
end

#average energy 
@average_energy GPTransition (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateGaussianDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::Wishart, meta::GPTransitionMeta,) = begin
    μ_y, Σ_y = mean_cov(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W_bar = mean(q_w) 
    E_logW = mean(logdet,q_w)
    Xu = getInducingInput(meta)
    Kuu_inverse = getinverseKuu(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    D = length(μ_y)
    M = length(Xu) #number of inducing points
    
    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ0 = approximate_expectation(GenUnscented(),q_in, (x) -> kernelmatrix(kernel, [x], [x]))[]
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D]  
    Σ_v = create_blockmatrix(Σ_v,D,M)
    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (Ψ0 - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    E = [getindex(Ψ1 * i,1) for i in μ_v]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_v[i]*μ_v[j]' + Σ_v[i,j]) * Ψ2)
        end
    end
    I2 = μ_y * μ_y' + Σ_y - μ_y * E' - E * μ_y' + Ψ_5 

    return 0.5*tr(W_bar*I1) + 0.5*D*log(2π) - 0.5*E_logW + 0.5 * tr(W_bar * I2)
end

@average_energy GPTransition (q_out::PointMass, q_in::MultivariateGaussianDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::Wishart, meta::GPTransitionMeta,) = begin
    μ_y = mean(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W_bar = mean(q_w) 
    E_logW = mean(logdet,q_w)
    Xu = getInducingInput(meta)
    Kuu_inverse = getinverseKuu(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    D = length(μ_y)
    M = length(Xu) #number of inducing points
    
    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ0 = approximate_expectation(GenUnscented(),q_in, (x) -> kernelmatrix(kernel, [x], [x]))[]
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D]  
    Σ_v = create_blockmatrix(Σ_v,D,M)
    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (Ψ0 - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    E = [getindex(Ψ1 * i,1) for i in μ_v]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_v[i]*μ_v[j]' + Σ_v[i,j]) * Ψ2)
        end
    end
    I2 = μ_y * μ_y' - μ_y * E' - E * μ_y' + Ψ_5 

    return 0.5*tr(W_bar*I1) + 0.5*D*log(2π) - 0.5*E_logW + 0.5 * tr(W_bar * I2)
end

@average_energy GPTransition (q_out::PointMass, q_in::MultivariateGaussianDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::PointMass, meta::GPTransitionMeta,) = begin
    μ_y = mean(q_out)
    μ_v, Σ_v = mean_cov(q_v)
    W_bar = mean(q_w) 
    E_logW = log(det(W_bar))
    Xu = getInducingInput(meta)
    Kuu_inverse = getinverseKuu(meta)
    cache = getGPCache(meta)
    kernel = getKernel(meta)
    D = length(μ_y)
    M = length(Xu) #number of inducing points
    
    Ψ1 = getcache(cache,(:Ψ1, M))
    Ψ2 = getcache(cache,(:Ψ2, (M,M)))
    Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
    Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
    Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 
    # Ψ0 = approximate_expectation(GenUnscented(),q_in, (x) -> kernelmatrix(kernel, [x], [x]))[]
    # Ψ1 = approximate_expectation(GenUnscented(), q_in,(x) -> kernelmatrix(kernel, [x], Xu))
    # Ψ2 = approximate_expectation(GenUnscented(), q_in, (x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu)) + 1e-7*I 
    μ_v = [view(μ_v,i:i+M-1) for i=1:M:M*D]  
    Σ_v = create_blockmatrix(Σ_v,D,M)
    trace_A = mul_A_B!(cache,Kuu_inverse,Ψ2,M) |> tr #trA =tr(Kuu_inverse * Ψ2)
    I1 = (Ψ0 - trace_A) .* C # kron(C, getindex(Ψ0,1) - trace_A)

    E = [getindex(Ψ1 * i,1) for i in μ_v]
    Ψ_5 = getcache(cache,(:Ψ_5,(D,D)))
    for j=1:D
        for i=1:D
            Ψ_5[i,j] = tr((μ_v[i]*μ_v[j]' + Σ_v[i,j]) * Ψ2)
        end
    end
    I2 = μ_y * μ_y' - μ_y * E' - E * μ_y' + Ψ_5 

    return 0.5*tr(W_bar*I1) + 0.5*D*log(2π) - 0.5*E_logW + 0.5 * tr(W_bar * I2)
end
# @average_energy GPTransition (q_out::MultivariateNormalDistributionsFamily, q_in::MultivariateGaussianDistributionsFamily, q_v::MultivariateNormalDistributionsFamily, q_w::PointMass, meta::GPTransitionMeta,) = begin
#     μ_y, Σ_y = mean_cov(q_out)
#     μ_v, Σ_v = mean_cov(q_v)
#     γ_bar = mean(q_w) 
#     E_logγ = log(γ_bar)
#     Xu = getInducingInput(meta)
#     Kuu_inverse = getinverseKuu(meta)
#     cache = getGPCache(meta)
#     D = length(μ_y)
#     M = length(Xu) #number of inducing points
    
#     Ψ0 = getcache(cache,(:Ψ0,(1,1)))
#     Ψ1 = getcache(cache,(:Ψ1,(1,M)))
#     Ψ2 = getcache(cache,(:Ψ2, (M,M)))
#     Ψ0 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], [x]),q_in)[]
#     Ψ1 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, [x], Xu),q_in)
#     Ψ2 = approximate_kernel_expectation(srcubature(),(x) -> kernelmatrix(kernel, Xu, [x]) * kernelmatrix(kernel, [x], Xu), q_in) 

#     I1 = D * (Ψ0 - tr(Kuu_inverse*Ψ2))
#     I2 = μ_y' * μ_y + tr(Σ_y) - 2*μ_y'*kron(C,Ψ1)*μ_v + tr((μ_v * μ_v' + Σ_v)*kron(C,Ψ2)) 
#     return 0.5*γ_bar*(I1 + I2) + 0.5*D*(log(2π) - E_logγ)
# end
```

## GPLVM model

```julia
@model function GPLVM(n, d_in, d_out, Kuu_inverse_big)
    #n: data length 
    #d_in: dimensionality of latent variables 
    #d_out: dimensionality of observations 
    y = datavar(Vector{Float64}, n)
    x = randomvar(n)

    v ~ MvNormalMeanCovariance(zeros(size(Kuu_inverse_big,1)), Kuu_inverse_big)
    # W ~ Wishart(1e2,0.1diageye(d_out))
    # W ~ GammaShapeRate(0.01,0.01)
    
    for i=1:n
        x[i] ~ MvNormalMeanCovariance(zeros(d_in), 50*diageye(d_in))
        y[i] ~ GPTransition(x[i],v,50*diageye(d_out)) where { q = MeanField() }  #mean-field assumption
    end
end

@meta function GPLVM_meta(Xu,Kuu_inverse,C,kernel,cache)
    GPTransition() -> GPTransitionMeta(Xu,Kuu_inverse,C,kernel,cache)
end
```

## Configure GP

```julia
Random.seed!(12)
M = 25
# Xu = [rand(Uniform(-3,3),2) for i=1:M];
Xu = [randn(2) for i=1:M]
# kernel_gp   =  1. * with_lengthscale(SEKernel(),[1., 1.]) 
kernel_gp   =  1. * with_lengthscale(SEKernel(),θ_optimal) 
meanfunc = (x) -> 0.0
Kuu = kernelmatrix(kernel_gp,Xu,Xu) + 1e-12*I;
Kuu_inverse = inv(Kuu) + 1e-12*I;
C = diageye(D_out)  #coregionalization matrix 
;
```

```julia
kernel_gp_param = (θ) -> with_lengthscale(SEKernel(),[θ[1],θ[2]]); 
kernelmatrix_param = (θ) -> kernelmatrix(kernel_gp_param(θ),Xu,Xu);
function my_free_energy(θ;qv=qv,qx=qx)
    my_Kuu = kernelmatrix_param(θ)
    my_Kuu_inverse = inv(my_Kuu)
    my_kernel = kernel_gp_param(θ)
    
    infer_result = infer(
        model = GPLVM(length(data),D_in,D_out,kron(C,my_Kuu_inverse)),
        iterations = 1,
        data = (y = data,),
        initmarginals = (v = qv, x = qx,),
        meta = GPLVM_meta(Xu, my_Kuu_inverse,C,my_kernel,GPCache()),
        free_energy = true,
    )
    return infer_result.free_energy[end]
end
function my_marginal(θ, qv, qx)
    my_Kuu = kernelmatrix_param(θ)
    my_Kuu_inverse = inv(my_Kuu)
    my_kernel = kernel_gp_param(θ)
    
    infer_result = infer(
        model = GPLVM(length(data),D_in,D_out,kron(C,my_Kuu_inverse)),
        iterations = 1,
        data = (y = data,),
        initmarginals = (v = qv, x = qx,),
        meta = GPLVM_meta(Xu, my_Kuu_inverse,C,my_kernel,GPCache()),
        returnvars = (x = KeepLast(), v = KeepLast())
    )
    qx = infer_result_gp.posteriors[:x];
    qv = infer_result_gp.posteriors[:v];
    return qx, qv 
end
function learn_params(θ,niter)
    θ_optimal = θ
    qv_current = MvNormalMeanCovariance(zeros(size(kron(C,Kuu),1)), kron(C,Kuu_inverse))
    qx_current = MvNormalMeanCovariance(zeros(D_in), 50*diageye(D_in))
    for i=1:niter
        res = optimize((x) -> my_free_energy(x;qv = qv_current,qx = qx_current),θ_optimal;iterations=1)
        θ_optimal = res.minimizer
        qx_current, qv_current = my_marginal(θ_optimal,qv_current, qx_current)
    end
    return θ_optimal, qx_current, qv_current
end
```
